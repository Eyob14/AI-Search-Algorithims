### Initialize imports

### Scatter graphs depicting time taken and path length in graph traversal algorithms for the romanian cities

```python
#first the csv file containig the results must be generated before running this code block
soln_Len=pd.read_csv(f"city_solnLength_test.csv")
soln_Len=soln_Len.groupby('path').mean()
soln_Len=soln_Len.reset_index()
soln_Len=soln_Len.drop(["path"],axis=1)
time_graph=pd.read_csv(f"city_time_test.csv")
time_graph=time_graph.groupby('path').mean()
time_graph=time_graph.reset_index()
time_graph=time_graph.drop(["path"],axis=1)
soln_Len=new_df = pd.DataFrame(soln_Len.mean()).T
time_graph=pd.DataFrame(time_graph.mean()).T
plt.figure(figsize=(18,10),dpi=200)
plt.title(f"Scatter graph for performance of different algorithms in the romanian cities graph",fontsize=25)
for algorithm in soln_Len:
    plt.scatter(time_graph[algorithm], soln_Len[algorithm],label=algorithm,s=soln_Len[algorithm]*50)
plt.xlabel('Time',fontsize=32)
plt.ylabel('Solution length',fontsize=32)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5),fontsize=16)
plt.show()
```

### Scatter graphs depicting time taken and path length in graph traversal algorithms for each of the 16 random graphs

```python
#first the csv file containig the results must be generated before running this code block
nodes=[10,20,30,40]
probs=[0.2,0.4,0.6,0.8]
for node in nodes:
    for prob in probs:
        soln_Len=pd.read_csv(f"{node} nodes {prob}_randomNode_solnLength_test.csv")
        soln_Len=soln_Len.groupby('path').mean()
        soln_Len=soln_Len.reset_index()
        soln_Len=soln_Len.drop(["path"],axis=1)
        time_graph=pd.read_csv(f"{node} nodes {prob}_randomNode_time_test.csv")
        time_graph=time_graph.groupby('path').mean()
        time_graph=time_graph.reset_index()
        time_graph=time_graph.drop(["path"],axis=1)
        soln_Len=new_df = pd.DataFrame(soln_Len.mean()).T
        time_graph=pd.DataFrame(time_graph.mean()).T
        plt.figure(figsize=(18,10),dpi=200)
        plt.title(f"Scatter graph representation for graph with {node} nodes with probability of edges {prob}",fontsize=20)
        for algorithm in soln_Len:
            plt.scatter(time_graph[algorithm], soln_Len[algorithm],label=algorithm,s=soln_Len[algorithm]*50)
        plt.xlabel('Time',fontsize=32)
        plt.ylabel('Solution length',fontsize=32)
        plt.legend(loc='center left', bbox_to_anchor=(1, 0.5),fontsize=16)
        plt.show()
```

### Bar graph for the centralities test

```python
centrality_dict={
    "degree_centrality":centrality.degree_centrality(graph),
    "closeness_centrality":centrality.closeness_centrality(graph),
    "betweeness_centrality":centrality.betweeness_centrality(graph),
    "pageRank_centrality":centrality.pageRank_centrality(graph),
    "eigenvector_centrality":centrality.eigenvector_centrality(graph),
    "katz_centrality":centrality.katz_centrality(graph)
}
for key in centrality_dict.keys():
    x=list(centrality_dict[key].keys())
    y=list(centrality_dict[key].values())
    plt.figure(figsize=(16,12))
    fig, ax = plt.subplots()
    title=key.replace("_"," ")
    plt.title(f"Bar Graph for {title}")
    plt.xticks(rotation=90)
    ax.bar(x, y)
    plt.show()
```
